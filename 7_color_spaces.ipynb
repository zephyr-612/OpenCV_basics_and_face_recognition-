```python
import cv2 as cv

img = cv.imread("H:\Code\Py\KIIT ML\OpenCV\Photos\park.jpg")
cv.imshow("img", img)

# BGR to Grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("gray", gray)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)  # hue saturation value
cv.imshow("HSV", hsv)

# BGR to l*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow("LAB", lab)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
plt.show()


cv.waitKey(0)
cv.destroyAllWindows()
```

```python
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow("RGB", rgb)

plt.imshow(rgb)
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# hsv to bgr
bgr = cv.cvtColor(hsv, cv.COLOR_HSV2BGR)
cv.imshow("BGR", bgr)

cv.waitKey(0)
cv.destroyAllWindows()
```
