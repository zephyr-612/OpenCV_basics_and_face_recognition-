```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("H:\Code\Py\KIIT ML\OpenCV\Photos\cats.jpg")
cv.imshow("Cats", img)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow("gray", gray)


cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Grayscale histogram
gray_hist = cv.calcHist([gray], [0], None, [256], [0, 256])
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels")
plt.plot(gray_hist)
plt.xlim([0, 256])
plt.show()
```

```python
img2 = cv.imread("H:\Code\Py\KIIT ML\OpenCV\Photos\cats 2.jpg")
gray2 = cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
gray2_hist = cv.calcHist([gray2], [0], None, [256], [0,256])
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(gray2_hist)
plt.xlim([0, 256])
plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np 
blank = np.zeros(img2.shape[:2], dtype = 'uint8')
circle = cv.circle(blank, (img2.shape[1]//2, img2.shape[0]//2), 100, 255, -1)
mask = cv.bitwise_and(gray2, gray2, mask = circle)
gray2_hist = cv.calcHist([gray2], [0], mask, [256], [0,256])
cv.imshow("Mask", mask)
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(gray2_hist)
plt.xlim([0, 256])
plt.show()


cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure()
plt.title("Color Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels")
colors =('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
plt.show()    

cv.waitKey(0)
cv.destroyAllWindows()    
```

```python
# Assuming img2 and circle are properly defined
# ...

# Ensure that the mask has the same dimensions as the image
if img2.shape[:2] != circle.shape[:2]:
    raise ValueError("Image and mask dimensions do not match")

mask2 = cv.bitwise_and(img2, img2, mask=circle)
cv.imshow('mask', mask2)

plt.figure()
plt.title("Color Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels")
colors = ('b', 'g', 'r')

for i, col in enumerate(colors):
    color_hist = cv.calcHist([img2], [i], mask2[:, :, i], [256], [0, 256])
    plt.plot(color_hist, color=col)
    plt.xlim([0, 256])

plt.show()

cv.waitKey(0)
cv.destroyAllWindows()
```
